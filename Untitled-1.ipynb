```python
import numpy as np
import matplotlib.pyplot as plt 
import keras
from keras.datasets import mnist

from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout 



```

```python

(X_train, y_train), (X_test, y_test) = mnist.load_data() # load our data
X_train.shape , y_train.shape, X_test.shape , y_test.shape   #for X_train ,The first dimension (60000) represents the number of training examples in the dataset.
                                                                     # (28) represents the height of each image in pixels.
                                                                     #(28) represents the width of each image in pixels.

```

```python


def plot_input_img(i): # for visualising our data
  plt.imshow(X_train[i], cmap='binary')
  plt.title(y_train[i])
  plt.axis('off')
  plt.show()


```

```python
for i in range(10):# top 10 data
  plot_input_img(i)
```

```python
# pre process the images

X_train = X_train.astype(np.float32)/255  # For normalising the image to [o,1] range 
X_test = X_test.astype(np.float32)/255

X_train = np.expand_dims(X_train, -1) # for expanding the dimension to (60000,28,28,1) . The new fourth dimension (1) represents the number of channels in the image.
                                      #In this case, there is only one channel, which means that the images are grayscale. 
                                      # If the images were in color, there would be three channels 
                                      #representing the red, green, and blue color channels. 255 means whitecolor and o means black and in b?w shades of grey s
X_test = np.expand_dims(X_test, -1)

# convert the classes to one hot vectors
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
```

```python
model= Sequential() # initialize our model 

model.add(Conv2D(32,(3,3), input_shape = (28,28,1) , activation= 'relu'))
model.add(MaxPool2D((2,2)))

model.add(Conv2D(64,(3,3),  activation= 'relu'))
model.add(MaxPool2D((2,2)))

model.add(Flatten())
model.add(Dropout(0.25))
model.add(Dense(10, activation="softmax")) # i have here 10 neuron because there is 1o no.of classes(0-9)
model.summary()


           
                      
```

```python
model.compile(optimizer= 'adam', loss = keras.losses.categorical_crossentropy , metrics= ['accuracy'])
```

```python
#Callbacks
from keras.callbacks import EarlyStopping, ModelCheckpoint
# Earlystopping
es = EarlyStopping(monitor='val_acc', min_delta= 0.01, patience= 4 , verbose=1)
#ModelCheckpoint
mc=ModelCheckpoint("./cse.h5", monitor="val_acc", verbose=1, save_best_only= False)
cb =[es,mc]
```

```python
his = model.fit(X_train, y_train, epochs= 5, validation_split=0.3, callbacks=cb)
```
